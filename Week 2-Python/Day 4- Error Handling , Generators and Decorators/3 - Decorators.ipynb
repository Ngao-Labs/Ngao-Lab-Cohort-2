Decorators can be thought of as functions which modify the *functionality* of another function. They help to make your code shorter and more "Pythonic".

```python
# Functions within functions

def hello():
    print(" This is where you can write Hello Function Code")
    def greet():
        return '\t This is the greet function inside the Hello Function'
    return greet
```

```python
func=hello()

print(func())
```

```python
# Functions as Arguments

def hello():
    return "Hello"

def greet(func):
    print("Greet Login goes here")
    print(func())

greet(hello)
```

```python
def new_decorator(func):
    def wrap_func():
        print('Code executing before the func')

        func()

        print("Code executing after func")
    return wrap_func
```

```python
def func_need_decorator():
    print("This function needs a decorator")
```

```python
# decorated_func= new_decorator(func_need_decorator)
@new_decorator
def func_need_decorator():
    print("This function needs a decorator")
```

```python
func_need_decorator()
```
