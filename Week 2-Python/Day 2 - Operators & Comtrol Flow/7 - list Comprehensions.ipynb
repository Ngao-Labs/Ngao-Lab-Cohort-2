# List Comprehensions

In addition to sequence operations and list methods, Python includes a more advanced operation called a list comprehension.

List comprehensions allow us to build out lists using a different notation. You can think of it as essentially a one line <code>for</code> loop built inside of brackets. For a simple example:


```python
list_characters=[]

for c in "Hello":
    list_characters.append(c)

list_characters
```

```python
list_characters=[c for c in "Hello"]

list_characters
```

```python
#List of all Squares from 0-10

Squares=[]

for num in range(0,11):
    Squares.append(num **2)

Squares
```

```python
squares=[ num**2  for num in range(0,11)]
Squares
```

```python
Squares=[]

for num in range(0,11):
    if num % 2 ==0:
        Squares.append(num **2)

Squares
```

```python
squares=[num**2  for num in range(0,11) if num %2 == 0 and num>5 ]
squares
```

```python
odd=[num**2  for num in range(0,11) if num %2 != 0  ]
print(odd)
```

```python
list1=[1,2,3]
list2=[10,20,30]

squares=[ x*y  for x in list1  for y in list2]
squares
```
