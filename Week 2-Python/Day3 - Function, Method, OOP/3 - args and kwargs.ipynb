```python
# Default parameters - they must come at the end

def calculate_tax( income, tax=30):
    return income * tax / 100

# print(calculate_tax(200000)) # will use default value for tax
# print(calculate_tax(200000, 16)) # will use 16 as tax


calculate_tax(200000, 16)
```

```python
def add_numbers( num1, num2, num3=0, num4=0):
    return num1 + num2 +num3 +num4


print(add_numbers(2,3))

print(add_numbers(2,3,5))


print(add_numbers(2,3,5,10))

# Write a function that adds the n numbers passed to it .
```

```python
def add_numbers(*args):
    print(sum(args))


add_numbers(2,3)
add_numbers(2,3,3,5,5,6,7,8,8,9,9,8,6,4)
# args are now a tuple
 
```

```python
def list_favourites(**kwargs):
    print(kwargs)

list_favourites(eat="Nyama choma", code="Python", play="Football", gaming="Minecraft")
```
