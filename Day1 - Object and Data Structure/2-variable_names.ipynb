# Variable
A variable is a container that stores data in your computer's memory. You give it a name and assign it a value, so you can use and change that data throughout your program.
To create a variable in Python, you simply choose a name for it and use the equals sign (`=`) to assign it a value. For example:


```python

age = 25
name = "Alice"
```

### Key Points

No declaration needed - just assign and go
Names are case-sensitive - Age and age are different
Values can change:


The names you use when creating these labels need to follow a few rules:

    1. Names can not start with a number.
    2. There can be no spaces in the name, use _ instead.
    3. Can't use any of these symbols :'",<>/?|\()!@#$%^&*~-+
    4. It's considered best practice (PEP8) that names are lowercase.
    5. Avoid using the characters 'l' (lowercase letter el), 'O' (uppercase letter oh), 
       or 'I' (uppercase letter eye) as single character variable names.
    6. Avoid using words that have special meaning in Python like "list" and "str"


Using variable names can be a very useful way to keep track of different variables in Python. For example:




````python
age= 25

age

print(age)
# print(id(age)) - Get Memory Id 
````
